```python
import pandas as pd
import os
```

```python
df = pd.read_csv("styles.csv", on_bad_lines='skip', engine='python')
```

```python
df.head()
```

```python
df['image'] = df.apply(lambda row: str(row['id']) + '.jpg', axis=1)
```

```python
# 1. Check for and handle rows with missing values in critical columns
critical_columns = ['id', 'masterCategory']
initial_rows = len(df)
df.dropna(subset=critical_columns, inplace=True)
rows_after_dropping_na = len(df)
if initial_rows != rows_after_dropping_na:
    print(f"Dropped {initial_rows - rows_after_dropping_na} rows with missing values in critical columns.")

# 2. Ensure 'id' column is in a consistent format (e.g., string)
try:
    df['id'] = df['id'].astype(str)
except ValueError:
    print("Warning: Could not convert 'id' column to string for some rows. Inspect data quality.")
    # Add more robust handling if needed, e.g., df = df[df['id'].astype(str, errors='coerce').notna()]

# 3. Re-create the 'image' column after cleaning
df['image'] = df.apply(lambda row: str(row['id']) + '.jpg', axis=1)

# 4. Filter out rows where the image file does not exist
existing_images = set(os.listdir('images'))
initial_rows_img_check = len(df)

# Store the rows that will be filtered out for inspection
invalid_filenames_df = df[~df['image'].isin(existing_images)].copy() # Use a different variable name

# Update df to only contain rows with existing images
df = df[df['image'].isin(existing_images)].copy() # Use .copy() to avoid SettingWithCopyWarning later

rows_after_img_check = len(df)

if initial_rows_img_check != rows_after_img_check:
     print(f"Dropped {initial_rows_img_check - rows_after_img_check} rows where the image file was not found.")
     print(f"Number of invalid filenames found during initial check: {len(invalid_filenames_df)}")
     # Print the filenames that were filtered out in this step
     print("Filenames not found during initial check:", invalid_filenames_df['image'].tolist())
```

```python
df.head(10)
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.head(10)
```

Preparing the imageDataGenerator for training the model

```python
!pip install keras_preprocessing
from keras_preprocessing.image import ImageDataGenerator
batch_size = 256
```

```python
image_generator = ImageDataGenerator(
    validation_split=0.2
)
```

```python
training_generator = image_generator.flow_from_dataframe(
    dataframe=df,
    directory='images',
    x_col='image',
    y_col='masterCategory',
    target_size=(60, 80),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)
```

```python
validation_generator = image_generator.flow_from_dataframe(
    dataframe=df,
    directory='images',
    x_col='image',
    y_col='masterCategory',
    target_size=(60, 80),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)
```

```python
classes = len(training_generator.class_indices)
classes
```

Initialising the model to Neural Networks

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
#initialising our model
model = Sequential()
model.add(Conv2D(32,(3,3), input_shape = (60,80,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size=(3,3)))

model.add(Conv2D(32,(3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size=(3,3)))

model.add(Flatten())


model.add(Dense(units = 32, activation = 'relu'))
model.add(Dense(units = 64, activation = 'relu'))
model.add(Dense(units = 128, activation = 'relu'))
model.add(Dense(units = 256, activation = 'relu'))
model.add(Dense(units = 512, activation = 'relu'))

model.add(Dense(units = classes, activation = 'softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

FITTING DATA IN OUR NEURAL NETWORKS

```python
from math import ceil
```

```python
model.fit(
    x = training_generator,
    steps_per_epoch = ceil(0.8 * (df.shape[0] / batch_size)),

    validation_data = validation_generator,
    validation_steps = ceil(0.2 * (df.shape[0] / batch_size)),

    epochs = 10,
    verbose = 1
)

loss, acc = model.evaluate(validation_generator, steps=ceil(0.2 * (df.shape[0] /batch_size)))
print('Loss: %.3f' % loss)
print('Accuracy: %.3f' % acc)
print("\n%s: %.2f%%"% (model.metrics_names[1], acc * 100))
```

Saving the model for future use

```python
model.save('fashion_price_predictor.keras')
```
